```python
def remove_duplicate(params, grads):
    
    params, grads = params[:], grads[:]
    
    while True:
        find_flg = False
        L = len(params)
        
        for i in range(0, L -1):
            for j in range(i + 1, L):
                if params[i] is params[j]:
                    grads[i] += grads[j]
                    find_flg = True
                    params.pop(j)
                    grads.pop(j)
                elif params[i].ndim == 2 and params[j].ndim == 2 and \
                     params[i].T.shape == params[j].shape and np.all(params[i].T == params[j]):
                    grads[i] += grads[j].T
                    find_flg = True
                    params.pop(j)
                    grads.pop(j)
                if find_flg: break
            if find_flg: break

        if not find_flg: break

    return params, grads

def clip_grads(grads, max_norm):
    total_norm = 0
    for grad in grads:
        total_norm += np.sum(grad ** 2)
    total_norm = np.sqrt(total_norm)

    rate = max_norm / (total_norm + 1e-6)
    if rate < 1:
        for grad in grads:
            grad *= rate

def accuracy(model, X, D):
    Y = model.predict(X)
    Y = np.argmax(Y, axis=1)
    if D.ndim != 1 : D = np.argmax(D, axis=1)

    accuracy = np.sum(Y == D) / float(X.shape[0])
    return accuracy
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from network import OriginalNet
from optimizer import SGD, Adam

# 
iris = datasets.load_iris()

# 
train_indices = np.random.choice(len(iris.data), round(len(iris.data)*0.8), replace=False)
# 
test_indices = np.array(list(set(range(len(iris.data))) - set(train_indices)))

x_train = iris.data[train_indices]
x_test = iris.data[test_indices]
t_train = iris.target[train_indices]
t_test = iris.target[test_indices]

# 
def normalize_cols(m):
    col_max = m.max(axis=0)
    col_min = m.min(axis=0)
    return (m-col_min) / (col_max - col_min)

x_train = np.nan_to_num(normalize_cols(x_train))
x_test = np.nan_to_num(normalize_cols(x_test))

# 
max_epoch = 500
batch_size = 30
hidden_size = 10
output_size = 3
learning_rate = 0.01

# 
data_size = len(x_train)
max_iters = data_size // batch_size
total_loss = 0
loss_count = 0
loss_list = []
accuracies_train = []
accuracies_test = []

model = OriginalNet(input_size=x_train.shape[1], hidden_size=hidden_size, output_size=output_size)
optimizer = Adam(lr = learning_rate)

for epoch in range(max_epoch):
    # 
    idx = np.random.permutation(data_size)
    x_train = x_train[idx]
    t_train = t_train[idx]

    for iters in range(max_iters):
        batch_x = x_train[iters*batch_size:(iters+1)*batch_size]
        batch_t = t_train[iters*batch_size:(iters+1)*batch_size]

        loss = model.forward(batch_x, batch_t)
        model.backward()
        params, grads = remove_duplicate(model.params, model.grads)
#         if max_grad is not None:
#             clip_grads
        optimizer.update(model.params, model.grads)

        total_loss += loss
        loss_count += 1

        accr_train = accuracy(model, x_test, t_test)
        accr_test = accuracy(model, x_test, t_test)
   
        avg_loss = total_loss / loss_count
        print('| epoch %d |  iter %d / %d | loss %.2f' % (epoch + 1, iters + 1, max_iters, avg_loss))
        total_loss, loss_count = 0, 0
    
    accuracies_train.append(accr_train) 
    accuracies_test.append(accr_test)  
    loss_list.append(avg_loss)

# 
plt.plot(range(max_epoch), loss_list, label='train')
plt.title("loss")
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()

plt.plot(range(max_epoch), accuracies_train, label='train')
plt.plot(range(max_epoch), accuracies_test, label='test')
plt.legend(loc="lower right")
plt.title("accuracy")
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.ylim(0, 1.0)
plt.show()
```
