```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from network import OriginalNet
from optimizer import SGD

# 
iris = datasets.load_iris()

# 
train_indices = np.random.choice(len(iris.data), round(len(iris.data)*0.8), replace=False)
# 
test_indices = np.array(list(set(range(len(iris.data))) - set(train_indices)))

x_train = iris.data[train_indices]
x_test = iris.data[test_indices]
t_train = iris.target[train_indices]
t_test = iris.target[test_indices]

# 
def normalize_cols(m):
    col_max = m.max(axis=0)
    col_min = m.min(axis=0)
    return (m-col_min) / (col_max - col_min)

x_train = np.nan_to_num(normalize_cols(x_train))
x_test = np.nan_to_num(normalize_cols(x_test))

# 
max_epoch = 1000
batch_size = 30
hidden_size = 10
output_size = 3
learning_rate = 0.1

# 
data_size = len(x_train)
max_iters = data_size // batch_size
total_loss = 0
loss_count = 0
loss_list = []

model = OriginalNet(input_size=x_train.shape[1], hidden_size=hidden_size, output_size=output_size)
optimizer = SGD(lr = learning_rate)

for epoch in range(max_epoch):
    # 
    idx = np.random.permutation(data_size)
    x_train = x_train[idx]
    t_train = t_train[idx]

    for iters in range(max_iters):
        batch_x = x_train[iters*batch_size:(iters+1)*batch_size]
        batch_t = t_train[iters*batch_size:(iters+1)*batch_size]

        loss = model.forward(batch_x, batch_t)
        model.backward()
        optimizer.update(model.params, model.grads)

        total_loss += loss
        loss_count += 1

        avg_loss = total_loss / loss_count
        print('| epoch %d |  iter %d / %d | loss %.2f' % (epoch + 1, iters + 1, max_iters, avg_loss))
        loss_list.append(avg_loss)
        total_loss, loss_count = 0, 0

# 
plt.plot(np.arange(len(loss_list)), loss_list, label='train')
plt.xlabel('iterations')
plt.ylabel('loss')
plt.show()
```
